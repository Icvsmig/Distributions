```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

1. Creating a binomial distribution with n = 20 and p = 0.5

```python
binomial = np.random.binomial(20,0.5, 100)
plt.hist(binomial)
plt.show()
```

```python
#Computing mean and Standard deviation 
binomial.mean(), binomial.std()

```

```python
#Plot a histogram for rand1.
plt.hist(binomial, color='c')

# Add a vertical line at the mean.
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binomial.mean()-binomial.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# Making a standard normally distributed variable with 1000 observations, a mean of 5, and 
# a standard deviation of 0.5, and putting it in a data frame.
mean = 5
sd = 0.5
n = 1000

df = pd.DataFrame({'rand': np.random.normal(mean, sd, n)})

# Plotting the variables in the data frame (here, just the variable "rand") as a histogram.
df.hist()
# Inline printing the histogram
plt.show()
```
